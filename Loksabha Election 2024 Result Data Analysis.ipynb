```python
df.sort_values(by='Margin',ascending=False).head(10)
```

```python
colors = [
    'orange' if i == 'Bharatiya Janata Party' else
    'red' if i == 'Indian National Congress' else
    'green' if i == 'All India Trinamool Congress' else
    'gray'  # Default color if none match
    for i in df.sort_values(by='Margin', ascending=False).head(10)['Leading Party']
]
```

```python
plt.bar(df.sort_values(by='Margin',ascending=False).head(10)['Constituency'],df.sort_values(by='Margin',ascending=False).head(10)['Margin'],label=df.sort_values(by='Margin',ascending=False).head(10)['Leading Party'],color=colors)
plt.legend()
plt.xlabel('Constituency')
plt.ylabel('Margin')
plt.title('top 10 seats won by biggest margin in loksabha 2024')
plt.xticks(rotation=90)
plt.savefig('election.png')
plt.figure(figsize=(25, 25))
```

```python
plt.figure(figsize=(10,10))
plt.ylabel('Average margin(in wining)')
plt.title('AVERAGE MARGIN VS ALL WON PARTY')
df.groupby('Leading Party')['Margin'].mean().sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(10,10))
plt.ylabel('Total constituancy')
plt.title('TOTAL WON constituency VS ALL WON PARTY')
df['Leading Party'].value_counts().plot(kind='bar',log=True)
```

```python
plt.figure(figsize=(10,10))
plt.title('frequency count of party which are second place')
df['Trailing Party'].value_counts().plot(kind='barh',log=True)
plt.xlabel('Total constituancy')
```
